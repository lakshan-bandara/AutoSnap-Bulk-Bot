## 📸 Step 2: Capture Screenshots
Enter your URLs and configuration below.

```python
# AutoSnap Configuration
URL_List = "https://github.com" #@param {type:"string"}
Resolution = "1920x1080" #@param {type:"string"}
Save_Folder = "Screenshots_Project" #@param {type:"string"}
Delay = 2 #@param {type:"slider", min:0, max:10, step:1}
Full_Page = True #@param {type:"boolean"}

import os, time
import google_colab_selenium as gs
from selenium.webdriver.chrome.options import Options

width, height = Resolution.split('x')

output_path = f"/content/drive/MyDrive/{Save_Folder}"
os.makedirs(output_path, exist_ok=True)

urls = [u.strip() for u in URL_List.split(',') if u.strip()]

print(f"📂 Saving to: {output_path}")
print(f"📸 Capturing {len(urls)} websites...")

# Initialize Driver using google-colab-selenium (More stable for Colab)
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = gs.Chrome(options=chrome_options)
driver.set_window_size(int(width), int(height))

for i, url in enumerate(urls, 1):
    try:
        # Ensure URL starts with http/https
        if not url.startswith('http'):
            url = 'https://' + url
            
        print(f"[{i}/{len(urls)}] Processing: {url}")
        driver.get(url)
        time.sleep(Delay)
        
        if Full_Page:
            # Scroll down to trigger lazy loading
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)
            driver.execute_script("window.scrollTo(0, 0);")
            
            # Get full page height
            total_height = driver.execute_script("return document.body.parentNode.scrollHeight")
            driver.set_window_size(int(width), total_height)
            time.sleep(1)

        # Create filename from URL
        clean_name = url.replace('https://','').replace('http://','').replace('/','_').replace('.','_')
        filename = f"snap_{clean_name}_{int(time.time())}.png"
        filepath = os.path.join(output_path, filename)

        driver.save_screenshot(filepath)
        print(f"✅ Saved: {filename}")
        
        # Reset window size for the next one
        driver.set_window_size(int(width), int(height))

    except Exception as e:
        print(f"⚠️ Error capturing {url}: {e}")

driver.quit()
print('\n✨ All screenshots captured successfully!')
```
