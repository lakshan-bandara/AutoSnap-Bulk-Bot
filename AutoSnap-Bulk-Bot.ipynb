## 📸 Step 2: Capture ScreenshotsEnter URLs and configuration below.

```python
# AutoSnap Configuration
URL_List = "https://google.com, https://github.com" #@param {type:"string"}
Resolution = "1920x1080" #@param {type:"string"}
Save_Folder = "Screenshots_Project" #@param {type:"string"}
Delay = 2 #@param {type:"slider", min:0, max:10, step:1}

import os, time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

width,height = Resolution.split('x')

output_path = f"/content/drive/MyDrive/{Save_Folder}"
os.makedirs(output_path, exist_ok=True)

urls = [u.strip() for u in URL_List.split(',') if u.strip()]

print(f"📂 Saving to: {output_path}")
print(f"📸 Capturing {len(urls)} websites...")

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=chrome_options)
driver.set_window_size(int(width), int(height))

for i,url in enumerate(urls,1):
    print(f"[{i}/{len(urls)}] Processing: {url}")
    driver.get(url)
    time.sleep(Delay)

    filename = url.replace('https://','').replace('http://','').replace('/','') + '.png'
    filepath = os.path.join(output_path, filename)

    driver.save_screenshot(filepath)

driver.quit()

print('\n✨ All screenshots captured and saved to Google Drive!')
```
