## 📸 Step 2: Capture Screenshots
Enter your URLs and configuration below.

```python
#@title AutoSnap Configuration
URL_List = "https://google.com, https://github.com" #@param {type:"string"}
Resolution = "1920x1080" #@param {type:"string"}
Save_Folder = "Screenshots_Project" #@param {type:"string"}
Delay = 2 #@param {type:"slider", min:0, max:10, step:1}

import datetime

# Create directory
output_path = f"/content/drive/MyDrive/{Save_Folder}"
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Clean URLs
urls = [u.strip() for u in URL_List.split(',') if u.strip()]

print(f"📂 Saving to: {output_path}")
print(f"📸 Capturing {len(urls)} websites...")

for i, url in enumerate(urls, 1):
    print(f"[{i}/{len(urls)}] Processing: {url}")
    # Remove http/https and special chars for filename if needed, but pageres handles it
    !pageres {url} {Resolution} --delay={Delay} --filename="<%= date %>-<%= url %>-<%= size %>" --crop
    
    # Move newly created files to drive
    !mv *.png "{output_path}/"

print("\n✨ All screenshots captured and saved to Google Drive!")
```
